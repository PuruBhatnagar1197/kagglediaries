```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

```

```python
X_train = np.linspace(-1, 1, 20)
y_train = np.array([-0.6561 , -0.3099 , -0.59035, -0.50855, -0.285  ,
                    -0.2443 , -0.02445,  0.00135, -0.2006 ,  0.07475,
                    -0.1422 ,  0.06515,  0.15265,  0.3521 ,  0.28415,
                    0.5524 ,  0.23115,  0.20835, 0.4211,  0.60485])
```

```python
X_test = np.linspace(-1, 1, 20)
y_test = np.array([-0.69415, -0.451  , -0.43005, -0.4484 , -0.1475 ,
                   -0.5019 , -0.28055,  0.24595, -0.21425, -0.0286 ,
                   0.23415,  0.46575, 0.07955,  0.1973 ,  0.0719 ,
                   0.3639 ,  0.5536 ,  0.3365 , 0.50705,  0.33435])
```

```python
plt.scatter(X_train, y_train, c='red', label='Train')
plt.scatter(X_test, y_test, c='blue', label='Test')
plt.legend()
plt.show()
```

Regression Model

```python
model=Sequential()

model.add(Dense(10,activation='relu',input_dim=1))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
adm=Adam(learning_rate=0.1)
```

```python
model.compile(optimizer=adm,metrics=['mse'],loss='mse')
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=500)
```

```python
_,train_mse=model.evaluate(X_train,y_train)
_,test_mse=model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

#### now adding dropout

```python
model_2=Sequential()

model_2.add(Dense(10,activation='relu',input_dim=1))
model_2.add(Dropout(0.2))
model_2.add(Dense(10,activation='relu'))
model_2.add(Dropout(0.2))
model_2.add(Dense(1,activation='linear'))
```

```python
model_2.summary()
```

```python
adm2=Adam(learning_rate=0.1)
model_2.compile(optimizer=adm2,metrics=['mse'],loss='mse')
```

```python
model_2.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=500)
```
